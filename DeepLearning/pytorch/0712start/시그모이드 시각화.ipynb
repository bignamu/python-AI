```python
import numpy as np 
import matplotlib.pylab as plt
```

```python
# [-10 , 10]
t = np.linspace(-10, 10 , 100)

# 시그모이드 공식 
# Numpy np.exp() 함수는 밑(base) 이 자연상수 e 인 지수함로 변환 해줍니다.
sig = 1 / (1 + np.exp(-t))

# t 와 시그모이드 결과값을 plotting 
plt.figure(figsize=(9,3))
plt.plot([-10, 10], [0,0], "k-")
plt.plot([-10, 10], [0.5, 0.5], "k:") # y 축 기준 0.5 실선 생성 
plt.plot([-10, 10], [1, 1], "k:") # y 축 기준 1.0 실선 생성 
plt.plot([0,0], [-1.1, 1.1], "k-") # center 기준 0.0 선 생성
plt.plot(t, sig, "r-", linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$") # Text를 보여주기위한 용도
plt.xlabel("t") # x축 이름 
plt.legend(loc="upper left", fontsize=20)
plt.axis([-10, 10, -0.1, 1.1]) # 그래프 간격 조정 
plt.show()
```

```python
# w값이 변화에 따른 경사도 변화 
def simgoid(t) :
    return 1 / (1 + np.exp(-t))

x = np.arange(-5.0, 5.0 , 0.1)
print(x)
y1 = simgoid(0.5 * x)
y2 = simgoid(x)
y3 = simgoid(2*x)

print("\n y1 simgoid \n", y1)
print("\n y2 simgoid \n", y2)
print("\n y3 simgoid \n", y3)

plt.plot(x, y1, 'r', linestyle = '--') # w의 값이 0.5
plt.plot(x, y2, 'g') # w의 값이 1 일때 
plt.plot(x, y3, 'b', linestyle = '--') # w의 값이 2 
plt.plot([0,0],[1.0 , 0.0]) # 가운데 점선 추가 
plt.title("Sigmoid Function")
plt.show()
```

```python
# b 값의 변화에 따른 좌 우 이동 
def simgoid(t) :
    return 1 / (1 + np.exp(-t))

x = np.arange(-5.0, 5.0 , 0.1)

y1 = simgoid(x+0.5)
y2 = simgoid(x+1)
y3 = simgoid(x+1.5)

print("\n y1 simgoid \n", y1)
print("\n y2 simgoid \n", y2)
print("\n y3 simgoid \n", y3)

plt.plot(x, y1, 'r', linestyle = '--') # x + 0.5
plt.plot(x, y2, 'g') # x + 1
plt.plot(x, y3, 'b', linestyle = '--') # x + 1.5
plt.plot([0,0],[1.0 , 0.0]) # 가운데 점선 추가 
plt.title("Sigmoid Function")
plt.show()
```
